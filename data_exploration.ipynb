```python
filepath = 'openmars_my30_ls292_my30_ls310.nc'
df = xr.open_dataset(filepath)
df = df.to_dataframe()
```

```python
df = df.reset_index()
df.describe()
```

```python
df = df.query('lat > 7.4 & lat < 7.6 & lon > 130 & lon < 140')
```

```python
df.describe()
```

```python
main_dataset.describe()
```

```python
appended_df = main_dataset.append(df)
appended_df.describe()
```

```python
appended_df.to_parquet('main_dataset2.parquet')
```

```python
ax = main_dataset.plot.scatter(x='time', y='dustcol', c='temp', colormap='inferno', s=0.2)
# ax.set_facecolor("lightslategray")
```

```python
main_avg_dataset = main_dataset.groupby(['lon', 'lat', 'time']).mean().reset_index()
```
